## Transform

```python
education_map = {
     'illiterate': 0,
     'basic.4y': 4,
     'basic.6y': 6,
     'basic.9y': 9,
     'high.school': 12,
     'professional.course': 12,
     'university.degree': 16,
     'unknown': np.NaN,
}
education_map
```

```python
df['education_numeric'] = df["education"].map(education_map)
```
